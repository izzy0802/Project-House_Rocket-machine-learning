# Imports

```python
import pandas as pd
from sklearn import linear_model as lr
import numpy as np
```

# Loading Data

```python
pf_row = pd.read_csv('Documents/kc_house_data.csv')
```

```python
pf_row.head()
```

# Data Prepration

```python
# features
x_train = pf_row.drop(['price','date'],axis=1)

# response variable
y_train = pf_row['price'].copy()
```

```python
x_train.head()
```

```python
y_train.head()
```

# Model Training

```python
# model descripition

modelo = lr.LinearRegression()

#model training
modelo.fit(x_train,y_train)
```

```python
# prediction

pred = modelo.predict(x_train)
```

# Performances Metrics

```python
df1 = pf_row.copy()
```

```python
df1['prediction'] = pred
```

```python
df1[['id','price','prediction','error','error_bas','error_porc','error_porc_abs']].head()
```

```python
df1['error'] = df1['price'] - df1['prediction']
```

```python
# somar so valores positivos
df1['error_bas'] = np.abs(df1['error'])
#porcentual de erro
df1['error_porc'] = (df1['price'] - df1['prediction']) / df1['price']
#erro porcentual absoluto
df1['error_porc_abs'] = np.abs(df1['error_porc'])
```

```python
#media
mae = np.mean(df1['error_bas'])
print(f'MAE: {mae}')
#media porcentual absoluto
mape = np.mean(df1['error_porc_abs'])
print(f'MAPE: {mape}')
```
